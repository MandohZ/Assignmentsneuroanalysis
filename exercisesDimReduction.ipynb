TSNE is a way to reduce a dataset with high dimensionality to an interpretable 2D map. The tunable parameter perplexity, one of the most important parameters, balances attention between local and global aspects of the data. The parameter is a guess about the number of close neighbors each point has. It has a complex effect on the plots. Typical values range between 5 and 50, to get the optimal plot, multiple perplexities need to be analyzed. source:https://distill.pub/2016/misread-tsne/




Changing the perplexity:




According to the documentation, The default perplexity in sklearn manififold TSNE is 30. When datasets become larger, the perplexity needs to be increased to visualize the global structure.

I can see clearly that the global structure of the data changes, there seems to be more overlap between datapoints, and the geometry of the datapoints changes. Since the perplexity is higher, it shows the structure of the data after a longer runtime. in short, a higher perplexity means better global structure of the data on the tSNE plot.

Exaggeration increases the attractive forces between points and allows points to more around more freely, which allows them to find their neighbors more easily. Exagerration can also be used during optimization forming more densily packed clusters, making seperation more visible.
